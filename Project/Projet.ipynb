```python
# Importer les biblioth√®ques n√©cessaires
import pandas as pd
import glob
import os



# Nombre maximum de lignes √† afficher (None pour un affichage complet)
lignes_max_a_afficher = 100 

# Nombre maximum de colonnes √† afficher (None pour un affichage complet)
colonnes_max_a_afficher = 50
# -------------------------------------------------------------

# Application des param√®tres d'affichage √† pandas
pd.set_option('display.max_rows', lignes_max_a_afficher)
pd.set_option('display.max_columns', colonnes_max_a_afficher)

# --- LECTURE ET FUSION DES FICHIERS CSV ---
#
# Chemin du dossier contenant vos fichiers CSV.
#  '.' pour signifier "le dossier actuel".
chemin_dossier = '.' 
# ---------------------------------------------

# Construire le mod√®le de recherche pour trouver tous les fichiers .csv
pattern_fichiers = os.path.join(chemin_dossier, "*.csv")

# Utiliser glob pour obtenir la liste de tous les fichiers correspondant au mod√®le
liste_fichiers_csv = glob.glob(pattern_fichiers)

# V√©rifier si des fichiers ont √©t√© trouv√©s
if not liste_fichiers_csv:
    print(f" AUCUN fichier CSV trouv√© dans le dossier : '{os.path.abspath(chemin_dossier)}'")
else:
    print(f" {len(liste_fichiers_csv)} fichiers CSV trouv√©s. D√©but de la fusion...")
    
    liste_dataframes = []

    # Boucler sur chaque nom de fichier trouv√©
    for fichier in liste_fichiers_csv:
        try:
            df_temp = pd.read_csv(fichier)
            liste_dataframes.append(df_temp)
            print(f"  -> '{fichier}' lu avec succ√®s ({len(df_temp)} lignes).")
        except Exception as e:
            print(f"  ->  Erreur lors de la lecture du fichier '{fichier}': {e}")

    # V√©rifier si au moins un DataFrame a √©t√© lu avec succ√®s
    if liste_dataframes:
        # Concat√©ner tous les DataFrames en un seul
        df_final = pd.concat(liste_dataframes, ignore_index=True)
        
        print("\n Fusion termin√©e avec succ√®s !")
        print(f"Le DataFrame final contient {df_final.shape[0]} lignes et {df_final.shape[1]} colonnes.")
        
        # Afficher le DataFrame final en respectant les param√®tres d√©finis plus haut
        print("\nAffichage du DataFrame fusionn√© :")
        display(df_final)
    else:
        print("\n Aucun fichier n'a pu √™tre lu correctement. La fusion est annul√©e.")
```

```python
# Importer les biblioth√®ques n√©cessaires
import pandas as pd
import glob
import os

# --- PARAM√àTRES D'AFFICHAGE ---
# Gardons la flexibilit√© pour le tableau final
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)
# --------------------------------

# --- LECTURE ET FUSION DES FICHIERS CSV ---
chemin_dossier = '.' 
pattern_fichiers = os.path.join(chemin_dossier, "*.csv")
liste_fichiers_csv = glob.glob(pattern_fichiers)
# ---------------------------------------------

if not liste_fichiers_csv:
    print(f"‚ùå AUCUN fichier CSV trouv√© dans le dossier : '{os.path.abspath(chemin_dossier)}'")
else:
    print(f"‚úÖ {len(liste_fichiers_csv)} fichiers CSV trouv√©s. Fusion en cours...")
    
    liste_dataframes = []
    for fichier in liste_fichiers_csv:
        try:
            df_temp = pd.read_csv(fichier)
            liste_dataframes.append(df_temp)
        except Exception as e:
            print(f"  -> ‚ö†Ô∏è Erreur lors de la lecture du fichier '{fichier}': {e}")

    if liste_dataframes:
        # Fusion des DataFrames
        df_final = pd.concat(liste_dataframes, ignore_index=True)
        print(f"\nüéâ Fusion termin√©e ! Le DataFrame contient {df_final.shape[0]} lignes et {df_final.shape[1]} colonnes.")
        
        # ======================================================================
        # √âTAPE 1 : AFFICHER LA LISTE DES COLONNES DISPONIBLES
        # ======================================================================
        print("\n" + "="*50)
        print("üïµÔ∏è Voici la liste des colonnes disponibles :")
        print("="*50)
        
        # On r√©cup√®re les noms de colonnes sous forme de liste
        liste_colonnes = df_final.columns.tolist()
        
        # On les affiche de mani√®re lisible
        for nom_colonne in liste_colonnes:
            print(f"- {nom_colonne}")
            
        print("="*50)
        
        # ======================================================================
        # √âTAPE 2 : S√âLECTIONNER ET AFFICHER LES COLONNES DE VOTRE CHOIX
        # ======================================================================
        print("\n‚úèÔ∏è MODIFIEZ LA LISTE CI-DESSOUS POUR CHOISIR VOS COLONNES.")
        
        # --- √Ä MODIFIER ---
        # Copiez-collez ici les noms exacts des colonnes que vous voulez voir.
        # Attention √† la casse (majuscules/minuscules) et aux espaces !
        colonnes_a_afficher = [
            'Nom_de_colonne_1',   # Exemple
            'Nom_de_colonne_3',   # Exemple
            'Autre_colonne'       # Exemple
        ]
        # ------------------
        
        try:
            # On cr√©e un nouveau DataFrame qui est un sous-ensemble du premier
            df_selection = df_final[colonnes_a_afficher]
            
            print(f"\n‚úÖ Affichage personnalis√© avec les {len(df_selection.columns)} colonnes s√©lectionn√©es :")
            display(df_selection)
            
        except KeyError as e:
            print(f"\n‚ùå ERREUR : Une des colonnes que vous avez choisies n'existe pas : {e}")
            print("Veuillez v√©rifier l'orthographe et la casse dans votre liste `colonnes_a_afficher`.")
        except Exception as e:
            print(f"\n‚ùå Une erreur inattendue est survenue : {e}")

    else:
        print("\n‚ùå Aucun fichier n'a pu √™tre lu correctement. La fusion est annul√©e.")
```
