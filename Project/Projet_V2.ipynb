```python
# --- 1. Cargement des donn√©es et Affichage

import pandas as pd
import glob
import os

# --- Affichage ---
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)

# --- Extraction des fichiers .csv ---
chemin_dossier = '.'
pattern_fichiers = os.path.join(chemin_dossier, "*.csv")
liste_fichiers_csv = glob.glob(pattern_fichiers)

if not liste_fichiers_csv:
    print(f" Aucun fichier CSV trouv√© dans le dossier : {os.path.abspath(chemin_dossier)}")

else:
    print(f" {len(liste_fichiers_csv)} fichiers CSV trouv√©s. Chargement en cours...")

    # --- Lecture et fusion ---
    liste_dataframes = []
    for fichier in liste_fichiers_csv:
        df_temp = pd.read_csv(fichier)
        liste_dataframes.append(df_temp)
        print(f"  ‚Üí {os.path.basename(fichier)} : {len(df_temp)} lignes")

    # --- Fusion compl√®te ---
    df = pd.concat(liste_dataframes, ignore_index=True)
    print(f"\n Fusion termin√©e ! Le DataFrame final contient {df.shape[0]} lignes et {df.shape[1]} colonnes.\n")

    # --- Normalisation des noms de colonnes ---
    df.columns = (
        df.columns
        .str.strip()           # Supprime espaces autour
        .str.replace(' ', '_') # Remplace espaces internes par underscores
        .str.lower()           # Met tout en minuscules
    )

   # print(" Colonnes disponibles (apr√®s nettoyage des noms) :")
   # print(df.columns.tolist())

    # --- V√©rification de la colonne cible ---
    if 'label' in df.columns:
        print("\n Colonne cible d√©tect√©e : 'label'")
        print("Valeurs uniques :", df['label'].unique())
    else:
        print("\n Aucune colonne 'label' d√©tect√©e.")
        
    
    # --- Aper√ßu ---
    print("\n Aper√ßu du DataFrame fusionn√© :")
    display(df.head())
```

```python
# --- 2. Nettoyage des donn√©es: Duplicats et Valeurs infinies ---

# Suppression des doublons
avant = df.shape[0]
df = df.drop_duplicates()
apres = df.shape[0]
print(f" {avant - apres} doublons supprim√©s ({apres} lignes restantes).")

# Remplacer les valeurs infinies par NaN
df.replace([float('inf'), float('-inf')], pd.NA, inplace=True)
print(" Valeurs infinies remplac√©es par NaN.")
```

```python
# --- 3. Gestion des valeurs manquantes ---

# Compter les valeurs manquantes
print(" Nombre de valeurs manquantes par colonne :")
display(df.isna().sum())

# Option 1 : Remplacer les NaN par la moyenne pour les colonnes num√©riques
colonnes_numeriques = df.select_dtypes(include=['number']).columns
df[colonnes_numeriques] = df[colonnes_numeriques].apply(lambda x: x.fillna(x.mean()))

# Option 2 : Remplacer les NaN par la m√©diane (plus robuste)
# df[colonnes_numeriques] = df[colonnes_numeriques].apply(lambda x: x.fillna(x.median()))

print(" Valeurs manquantes remplac√©es par la moyenne des colonnes num√©riques.")
```

```python
# --- 4. Supression des collonnes inutiles ---

# Identifier les colonnes avec une seule valeur unique
colonnes_uniques = [col for col in df.columns if df[col].nunique() == 1]

if colonnes_uniques:
    print(f" Colonnes supprim√©es car sans variance ({len(colonnes_uniques)} au total) :")
    print(colonnes_uniques)
    df.drop(columns=colonnes_uniques, inplace=True)
else:
    print(" Aucune colonne inutile d√©tect√©e (toutes ont une variance).")

print(f"\n Dimensions finales du DataFrame : {df.shape[0]} lignes, {df.shape[1]} colonnes.")
```

```python
# --- 5. Labeling des donn√©es ---

# V√©rification de l'existence de 'label'
if 'label' not in df.columns:
    raise KeyError(" La colonne 'label' n'existe pas dans le DataFrame.")

# Aper√ßu des labels uniques avant transformation
print(" Labels uniques avant mapping :")
print(df['label'].unique())

# --- MAPPING ---
# On regroupe plusieurs attaques sous des cat√©gories plus g√©n√©rales

mapping_label = {
    # 0 ‚Üí BENIGN
    'BENIGN': 0,
    
    # 1 ‚Üí Brute Force (SSH, FTP, etc.)
    'SSH-Bruteforce': 1,
    'FTP-BruteForce': 1,
    'Brute Force': 1,
    
    # 2 ‚Üí DDoS (Distributed Denial of Service)
    'DDoS': 2,
    'DDOS attack-HOIC': 2,
    'DDOS attack-LOIC-UDP': 2,
    'DDOS attack-LOIC-HTTP': 2,
    
    # 3 ‚Üí DoS (Denial of Service)
    'DoS GoldenEye': 3,
    'DoS Hulk': 3,
    'DoS Slowhttptest': 3,
    'DoS slowloris': 3,
    'DoS': 3,
    
    # 4 ‚Üí Heartbleed
    'Heartbleed': 4,
    
    # 5 ‚Üí Infiltration
    'Infiltration': 5,
    
    # 6 ‚Üí Port Scan
    'PortScan': 6,
    'Port Scan': 6,
    
    # 7 ‚Üí Web Attack (regroupe plusieurs sous-types)
    'Web Attack ‚Äì Brute Force': 7,
    'Web Attack ‚Äì XSS': 7,
    'Web Attack ‚Äì Sql Injection': 7,
    'WebAttack': 7
}

# --- APPLICATION DU MAPPING ---
df['Attack Number'] = df['label'].map(mapping_label)

# V√©rifions si des labels n‚Äôont pas trouv√© de correspondance
labels_non_map = df[df['Attack Number'].isna()]['label'].unique()
if len(labels_non_map) > 0:
    print("\n Attention : certains labels n‚Äôont pas √©t√© mapp√©s (valeurs NaN) :")
    print(labels_non_map)
else:
    print("\n Tous les labels ont √©t√© mapp√©s avec succ√®s.")

# --- AFFICHAGE DU R√âSULTAT FINAL ---
print("\n Distribution des classes apr√®s mapping :")
display(df['Attack Number'].value_counts().sort_index())

# Exemple de correspondance claire
print("\n Mapping appliqu√© :")
for label, code in mapping_label.items():
    print(f"{code}: {label}")
```

```python
# --- 4. Equilibrage  ---

from sklearn.utils import resample

# V√©rifions que la colonne d‚Äôattaque ("attack_number") existe
if 'Attack Number' not in df.columns:
    raise KeyError(" La colonne 'attack_number' est introuvable.")

# --- √âtape 1 : Cr√©ation de la colonne binaire "attack_type" ---
# 0 ‚Üí BENIGN
# 1 ‚Üí Toute forme d‚Äôattaque
df['attack_type'] = df['Attack Number'].apply(lambda x: 0 if x == 0 else 1)

print(" Nouvelle colonne binaire 'attack_type' cr√©√©e :")
display(df['attack_type'].value_counts())

# --- √âtape 2 : V√©rification du d√©s√©quilibre ---
count_benign = df[df['attack_type'] == 0].shape[0]
count_attack = df[df['attack_type'] == 1].shape[0]

print(f"\n R√©partition initiale :")
print(f" - BENIGN (0) : {count_benign}")
print(f" - ATTACK (1) : {count_attack}")

# --- √âtape 3 : √âquilibrage du jeu de donn√©es ---
# D√©finir la taille cible pour chaque classe 
target_size = 10_000

# √âchantillonnage al√©atoire sans remplacement
benign_sample = resample(
    df[df['attack_type'] == 0],
    replace=False,
    n_samples=min(target_size, count_benign),
    random_state=42
)

attack_sample = resample(
    df[df['attack_type'] == 1],
    replace=False,
    n_samples=min(target_size, count_attack),
    random_state=42
)

# Combiner les deux sous-ensembles √©quilibr√©s
df_balanced = pd.concat([benign_sample, attack_sample], ignore_index=True)

print(f"\n Jeu de donn√©es √©quilibr√© : {df_balanced.shape[0]} lignes totales")
print("Distribution finale :")
display(df_balanced['attack_type'].value_counts())

# --- √âtape 4 : Aper√ßu rapide du DataFrame √©quilibr√© ---
display(df_balanced.head())
```

```python
# --- 5. Sauvegarde du DataSet propre ---

import os

# V√©rifier que df_balanced existe
if 'df_balanced' not in locals():
    raise NameError(" Le DataFrame 'df_balanced' n'existe pas.")

# --- √âtape 1 : Cr√©er un sous-√©chantillon pour les exp√©riences ---
# On prend ~15 000 lignes al√©atoirement
subset_size = 15_000
df_subset = df_balanced.sample(n=min(subset_size, len(df_balanced)), random_state=42)

print(f" Sous-√©chantillon cr√©√© : {df_subset.shape[0]} lignes s√©lectionn√©es pour les exp√©riences de labo.")

# --- √âtape 2 : Cr√©er le dossier de sortie (si non existant) ---
output_dir = "data"
os.makedirs(output_dir, exist_ok=True)

# --- √âtape 3 : Sauvegarde en Parquet  ---
parquet_path = os.path.join(output_dir, "data_processed.parquet")
df_subset.to_parquet(parquet_path, index=False)
print(f" Fichier Parquet sauvegard√© : {parquet_path}")

# --- √âtape 4 : Sauvegarde de secours en CSV ---
csv_path = os.path.join(output_dir, "data_processed.csv")
df_subset.to_csv(csv_path, index=False)
print(f" Fichier CSV sauvegard√© : {csv_path}")

# --- √âtape 5 : V√©rification rapide ---
print("\n V√©rification du DataFrame sauvegard√© :")
display(df_subset.head())
```
